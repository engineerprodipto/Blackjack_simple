```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
class Card:
    
    def __init__(self,suit,rank):
        
        self.suit=suit
        self.rank=rank
        #self.value=values[rank]
    
    def __str__(self):
        
        return self.rank + " of " + self.suit
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        deck_comp=''
        for cards in self.deck:
            deck_comp += '\n' + cards.__str__()
        return 'The deck has: '+ deck_comp

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop()
```

```python
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
        
    def add_card(self,card):
        #this card is gonna come from Deck.deal()
        self.cards.append(card)
        self.value+=values[card.rank]
        
        if card.rank == 'Ace':
            self.aces+=1
        
    
    def adjust_for_ace(self):
        while self.value>21 and self.aces>0:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips:
    
    def __init__(self, total=100):
        self.total = total  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True:
        
        try:
            chips.bet = int(input("How many chips would you like to bet? "))
        except:
            print("Sorry please provide an integer!")
        else:
            if chips.bet>chips.total:
                print("Sorry! You do not have enough chips! You have: {}".format(chips.total))
            else:
                break
    
```

```python
def hit(deck,hand):
    
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
    
```

```python
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    while True:
        x=input("Hit or Stand? H/S")
        
        if x[0].lower()=='h':
            hit(deck,hand)
            break
        elif x[0].lower() == 's':
            playing= False
            break
        else:
            print("Sorry, please try again! ")
            continue
        
```

```python
def show_some(player,dealer):
    
    print("\n Dealer's Hand: ")
    print("First card of dealer is hidden!")
    print(dealer.cards[1])
    
    print("\n Player's Hand: ")
    for card in player.cards:
        print(card)
    
def show_all(player,dealer):
    print("\n Dealer's Hand: ")
    for card in dealer.cards:
        print(card)
        
    print(f"\nValue of the dealer's hand is: {dealer.value}")
    
    print("\n Player's Hand: ")
    for card in player.cards:
        print(card)
        
    print(f"\nValue of the player's hand is: {player.value}")
    
```
